# Get titles of the news from Varzesh3

```python
from bs4 import BeautifulSoup
import requests
import csv
from urllib.request import urlopen as uReq
```

```python
csv_file =  open('varzesh3.csv' , 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['title' , 'lead' , 'content'])
```

```python
source = requests.get("http://www.varzesh3.com").text
soup = BeautifulSoup(source, 'lxml')
```

```python
tab = soup.findAll('div' , class_ ='tab-pane' , id = 'mostRecentNews')
links = tab[1].findAll('a')     #tab[1] is soccer related news, tab[0] other sports
```

```python
# list of the links of all recent soccer realted news
list_links = []
for item in links:
    website = "http://varzesh3.com"
    link = "/".join(item['href'].split('/')[1:3])
    full_address = website+"/"+link
    list_links.append(full_address)
```

```python

for address in list_links[0:2]:     #only try the first two lines for now
    source = requests.get(address).text
    soup = BeautifulSoup(source, 'lxml')
    try:
        title = soup.find('h1' , class_ = 'news-page--news-title').text
        lead = soup.find('p' , class_ = 'news-page--news-lead text-justify').text
        content = soup.find('div' , class_ = 'col-xs-12 news-page--news-text text-justify').text
    except:
        print('no text!')
```
