# Get titles of the news from Varzesh3

```python
from bs4 import BeautifulSoup
import requests
import csv
from urllib.request import urlopen as uReq
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
with open('varzesh3.csv' , 'w') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['title' , 'lead' , 'content'])
```

```python
source = requests.get("http://www.varzesh3.com").text
soup = BeautifulSoup(source, 'lxml')
```

```python
tab = soup.findAll('div' , class_ ='tab-pane' , id = 'mostRecentNews')
links = tab[1].findAll('a')     #tab[1] is soccer related news, tab[0] other sports
```

```python
# list of the links of all recent soccer realted news
list_links = []
for item in links:
    website = "http://varzesh3.com"
    link = "/".join(item['href'].split('/')[1:3])
    full_address = website+"/"+link
    list_links.append(full_address)
```

```python
with open('test.txt' , encoding = "utf_8"  , mode = 'w' , ) as file:
    pass
n = 1
for address in list_links[0:10]:     #only try the first two lines for now
    source = requests.get(address).text
    soup = BeautifulSoup(source, 'lxml')
    try:
        title = soup.find('h1' , class_ = 'news-page--news-title').text
        lead = soup.find('p' , class_ = 'news-page--news-lead text-justify').text
        content = soup.find('div' , class_ = 'col-xs-12 news-page--news-text text-justify').text
        with open('test.txt' ,encoding = "utf_8" ,  mode = 'a') as f:
            f.write(title)
            f.write(lead)
            f.write(content)
        with open('varzesh3.csv' , 'a') as csv_file:
            csv_writer = csv.writer(csv_file)
            csv_writer.writerow([title, lead, content])  
        print(f'This is news number {n}. \n')
        print(title , '\n' , lead , '\n', content , '\n')
        n += 1
    except:
        print('no text!')
```
